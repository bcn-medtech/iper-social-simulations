## Nodes

```python
with open('../shapefiles/Metro/tram_stations.json') as json_file:
    all_data = json.load(json_file)
    
data = all_data['data']['tram']
df = pd.DataFrame(data)
df['id'] = np.arange(df.shape[0])

gdf_tram_stations = gpd.GeoDataFrame( df, geometry= gpd.points_from_xy(df.lon, df.lat))
gdf_tram_stations.set_crs(epsg=ox.settings.default_crs[5:9], inplace=True)
gdf_tram_stations = ox.project_gdf(gdf_tram_stations)
gdf_tram_stations['y'] = gdf_tram_stations.geometry.y
gdf_tram_stations['x'] = gdf_tram_stations.geometry.x
gdf_tram_stations
```

## Edges

```python
with open('../shapefiles/Metro/tram_edges.json') as json_file:
    all_data = json.load(json_file)
    
data = all_data['data']['tram']
geometries = []
df_unique = pd.DataFrame(data)

# Create the BidirectionalGraph 
dest_bi_edges = []
orig_bi_edges = []
coord_bi_edges = []

for i in np.arange(len(df_unique)):
    dest_bi_edges.append(df_unique['origin'][i])
    orig_bi_edges.append(df_unique['destination'][i])
    coord_edges = df_unique['coordinates'][i]
    str_coords = coord_edges.split(',0')[:-1] # Separate the coord in packs of 2
    coord_bi_edges.append(list(reversed(str_coords)))

bi_edges_data = {'id': np.zeros(df_unique.shape[0]),
                'line': list(df_unique['line']),
                'origin': orig_bi_edges,
                'destination': dest_bi_edges,
                'layer': list(df_unique['layer']),
                'coordinates': coord_bi_edges}

df_bi = pd.DataFrame(bi_edges_data)
df = pd.concat([df_unique,df_bi])
df['id'] = np.arange(gdf_tram_stations.shape[0],gdf_tram_stations.shape[0]+df.shape[0])

# Create the geometric attribut for the unique edges
for edge_coords in df['coordinates'][:len(df_unique)]:
    str_coords = edge_coords.split(',0') # Separate the coord in packs of 2
    coords = [(*map(float,i.split(',')),) for i in str_coords[:-1]] ## convert string to float
    geometries.append(LineString(coords))   # Generate line strings

# Create the geometric attribut for the Bi edges   
for edge_coords in df['coordinates'][len(df_bi):len(df)]:
    coords = [(*map(float,i.split(',')),) for i in edge_coords] ## convert string to float
    geometries.append(LineString(coords))   # Generate line strings

# Convert the dataframe to geodataframe
gdf_tram_edges = gpd.GeoDataFrame(df, geometry= geometries)

# Set crs and project the graph. 
gdf_tram_edges.set_crs(epsg=ox.settings.default_crs[5:9], inplace=True)
gdf_tram_edges = ox.project_gdf(gdf_tram_edges)
```

```python
# Create u and v
u = []
v = []
for origin, line in zip(gdf_tram_edges['origin'], gdf_tram_edges['line']):
    node_id = gdf_tram_stations.loc[(gdf_tram_stations['name'] == origin)]['id'].item() 
    u.append(int(node_id))

for dest, line in zip(gdf_tram_edges['destination'], gdf_tram_edges['line']):
    node_id = gdf_tram_stations.loc[(gdf_tram_stations['name'] == dest)]['id'].item()
    v.append(int(node_id))

gdf_tram_edges['v'] = v
gdf_tram_edges['u'] = u
```

```python
gdf_tram_edges['key'] = np.zeros(len(gdf_tram_edges))
gdf_tram_edges['length']= gdf_tram_edges.geometry.length
```

```python
gdf_tram_stations = gdf_tram_stations.rename(columns={'id': 'osmid'})
gdf_tram_edges = gdf_tram_edges.rename(columns={'id': 'osmid'})
G = ox.graph_from_gdfs(gdf_tram_stations,gdf_tram_edges)
```

```python
ox.plot_graph(G)
```

```python
# find the shortest path (by distance) between these nodes then plot it
orig = list(G)[50]
dest = list(G)[30]

route = ox.shortest_path(G , orig, dest, weight='length')
fig, ax = ox.plot_graph_route(G , route, route_color='y', route_linewidth=6, node_size=0)
```

```python
# Save the graph 
ox.save_graphml(G, filepath="../BCNgraphs/tram.graphml")
```

## Nodes

```python
with open('../shapefiles/Metro/metro_stations.json') as json_file:
    all_data = json.load(json_file)

data = all_data['data']['metro']
df = pd.DataFrame(data)
df['id'] = np.arange(df.shape[0])

gdf_subway_stations = gpd.GeoDataFrame( df, geometry= gpd.points_from_xy(df.lon, df.lat))
gdf_subway_stations.set_crs(epsg=ox.settings.default_crs[5:9], inplace=True)
gdf_subway_stations = ox.project_gdf(gdf_subway_stations)
gdf_subway_stations['y'] = gdf_subway_stations.geometry.y
gdf_subway_stations['x'] = gdf_subway_stations.geometry.x
gdf_subway_stations
```

## Edges

```python
#subway_edges = requests.get("http://barcelonaapi.marcpous.com/metro/edges.json")
#with open('../shapefiles/Metro/metro_edges.json', 'wb') as f:
#    f.write(subway_edges.content)
```

```python
with open('../shapefiles/Metro/metro_edges.json') as json_file:
    all_data = json.load(json_file)

data = all_data['data']['metro']
geometries = []
df_unique = pd.DataFrame(data)

# Create the BidirectionalGraph 
dest_bi_edges = []
orig_bi_edges = []
coord_bi_edges = []

for i in np.arange(len(df_unique)):
    dest_bi_edges.append(df_unique['origin'][i])
    orig_bi_edges.append(df_unique['destination'][i])
    coord_edges = df_unique['coordinates'][i]
    str_coords = coord_edges.split(',0')[:-1] # Separate the coord in packs of 2
    coord_bi_edges.append(list(reversed(str_coords)))

bi_edges_data = {'id': np.zeros(df_unique.shape[0]),
                'line': list(df_unique['line']),
                'origin': orig_bi_edges,
                'destination': dest_bi_edges,
                'layer': list(df_unique['layer']),
                'coordinates': coord_bi_edges}

df_bi = pd.DataFrame(bi_edges_data)
df = pd.concat([df_unique,df_bi])
df['id'] = np.arange(gdf_subway_stations.shape[0], gdf_subway_stations.shape[0]+df.shape[0])

# Create the geometric attribut for the unique edges
for edge_coords in df['coordinates'][:len(df_unique)]:
    str_coords = edge_coords.split(',0') # Separate the coord in packs of 2
    coords = [(*map(float,i.split(',')),) for i in str_coords[:-1]] ## convert string to float
    geometries.append(LineString(coords))   # Generate line strings

# Create the geometric attribut for the Bi edges   
for edge_coords in df['coordinates'][len(df_bi):len(df)]:
    coords = [(*map(float,i.split(',')),) for i in edge_coords] ## convert string to float
    geometries.append(LineString(coords))   # Generate line strings
    
# Convert the dataframe to geodataframe
gdf_subway_edges = gpd.GeoDataFrame(df, geometry= geometries)

# Set crs and project the graph. 
gdf_subway_edges.set_crs(epsg=ox.settings.default_crs[5:9], inplace=True)
gdf_subway_edges = ox.project_gdf(gdf_subway_edges)
```

```python
# Create u and v
u = []
v = []
for origin, line in zip(gdf_subway_edges['origin'], gdf_subway_edges['line']):
    try:
        node_id = gdf_subway_stations.loc[(gdf_subway_stations['name'] == origin) & 
                                     (gdf_subway_stations['line'] == line)]['id'].item()
    except:
        print(origin)
        print(line)
        
    u.append(int(node_id))

for dest, line in zip(gdf_subway_edges['destination'], gdf_subway_edges['line']):
    try:
        node_id = gdf_subway_stations.loc[(gdf_subway_stations['name'] == dest) & 
                                         (gdf_subway_stations['line'] == line)]['id'].item()
    except:
        print(dest)
        print(line)   
    v.append(int(node_id))

gdf_subway_edges['v'] = v
gdf_subway_edges['u'] = u
```

```python
#Create edges between lines 
id_connection = []
line_connection = []
origin_connection = []
dest_connection = []
geometry_connection = []
u_connection = []
v_connection = []
element = 0

for index, row in gdf_subway_stations.iterrows():
    for index_compare, row_compare in gdf_subway_stations.iterrows():
        if index == index_compare:
            continue
        if row['name'] == row_compare['name']:
            id_connection.append(gdf_subway_edges.shape[0]+element)
            line_connection.append(row['line'] + '|' + row_compare['line'])
            origin_connection.append(row['name'])
            dest_connection.append(row_compare['name'])
            geometry_connection.append(LineString([row['geometry'],row_compare['geometry']]))
            u_connection.append(index)
            v_connection.append(index_compare)
            element += 1
            
        
connect_lines = {'id': id_connection,
                'line': line_connection,
                'origin': origin_connection,
                'destination': dest_connection,
                'geometry': geometry_connection,
                'u': u_connection,
                'v': v_connection}

df_connect = pd.DataFrame(connect_lines)
gdf_subway_edges = pd.concat([gdf_subway_edges,df_connect])
```

```python
gdf_subway_edges['key'] = np.zeros(len(gdf_subway_edges))
gdf_subway_edges['length']= gdf_subway_edges.geometry.length
```

```python
gdf_subway_stations = gdf_subway_stations.rename(columns={'id': 'osmid'})
gdf_subway_edges = gdf_subway_edges.rename(columns={'id': 'osmid'})
G = ox.graph_from_gdfs(gdf_subway_stations,gdf_subway_edges)
ox.save_graphml(G, filepath="../BCNgraphs/subway.graphml")
```

```python
ox.plot_graph(G)
```

```python
# find the shortest path (by distance) between these nodes then plot it
orig = list(G)[121]
dest = list(G)[100]

route = ox.shortest_path(G , orig, dest, weight='length')
print(route)
fig, ax = ox.plot_graph_route(G , route, route_color='r', route_linewidth=6, node_size=10)
```

## Edges

```python
with open('../shapefiles/Metro/renfe_edges.json') as json_file:
    all_data = json.load(json_file)
    
data = all_data['data']['renfe']
df = pd.DataFrame(data)
df = df.loc[df['type'] == 'RODALIES DE CATALUNYA-REGIONALS']
```

```python

for index, row in df.iterrows():
    longitude = []
    latitude = []
    coordinates = row['coordinates'].split(",0")[:-1]
    for coord in coordinates:
        lon, lat = coord.split(",")
        longitude.append(lon)
        latitude.append(lat) 

    if any(float(ele) >= northern_node for ele in latitude) | any(float(ele) <= southern_node for ele in latitude) | any(float(ele) >= eastern_node for ele in longitude) | any(float(ele) <= western_node for ele in longitude):
        df = df.drop(index,axis = 0)
        
df
```
